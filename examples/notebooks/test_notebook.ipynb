```python
from unittest.mock import patch
from policy_corpus_builder.adapters import non_eu

term = ["environment"]

# 1) Baseline: current behavior
df_baseline = non_eu.fetch_uk_documents(term, max_per_term=3, sleep_s=0)
print("baseline rows:", len(df_baseline))
print(df_baseline[["title", "url"]].head())

# 2) Invocation test:
# If this raises, then the DuckDuckGo fallback was actually reached.
try:
    with patch(
        "policy_corpus_builder.adapters.non_eu._fetch_uk_search_results_via_duckduckgo",
        side_effect=RuntimeError("DUCKDUCKGO_FALLBACK_WAS_CALLED"),
    ):
        df_test = non_eu.fetch_uk_documents(term, max_per_term=3, sleep_s=0)
    print("fallback was NOT called; rows:", len(df_test))
except RuntimeError as e:
    print("fallback invocation test:", e)

# 3) Necessity test:
# If rows drop to 0 here, then the fallback is not just called, it is required.
with patch(
    "policy_corpus_builder.adapters.non_eu._fetch_uk_search_results_via_duckduckgo",
    return_value=[],
):
    df_no_ddg = non_eu.fetch_uk_documents(term, max_per_term=3, sleep_s=0)

print("rows with DuckDuckGo disabled:", len(df_no_ddg))
if len(df_no_ddg):
    print(df_no_ddg[["title", "url"]].head())
```
