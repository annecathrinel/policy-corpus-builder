```python
import json
from pathlib import Path

repo_root = Path.cwd()
if repo_root.name == "notebooks":
    repo_root = repo_root.parent.parent

path = repo_root / "examples" / "outputs" / "non-eu-uk-example" / "documents.jsonl"
print(path)
print(path.exists())

rows = [json.loads(line) for line in path.read_text(encoding="utf-8").splitlines() if line.strip()]

print("records:", len(rows))
print("with full_text:", sum(bool(r.get("full_text")) for r in rows))
print("without full_text:", sum(not bool(r.get("full_text")) for r in rows))

for r in rows[:5]:
    print({
        "title": r.get("title"),
        "url": r.get("url"),
        "has_full_text": bool(r.get("full_text")),
    })
```
